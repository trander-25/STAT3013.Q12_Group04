Preprocessing Pipline
-

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), X_train.select_dtypes(exclude=['object']).columns),
        ('cat', OneHotEncoder(handle_unknown='ignore'), X_train.select_dtypes(include=['object']).columns)
    ]
)
```

Build MLP model
-

```python
mlp = MLPClassifier(hidden_layer_sizes=(100, 50),
                    activation='relu',
                    solver='adam',
                    learning_rate_init=0.001,
                    max_iter=1000,
                    random_state=42,
                   early_stopping=True)

model = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('classifier', mlp)
])
```

```python
model.fit(X_train, y_train)
```

Tune threshold to increase recall

```python
from sklearn.metrics import precision_score, recall_score

# Lấy xác suất của class 1 (churn)
y_proba = model.predict_proba(X_test)[:, 1]

thresholds = [0.5, 0.4, 0.3, 0.2]

for t in thresholds:
    y_pred = (y_proba >= t).astype(int)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    print(f"Threshold={t:.2f} | Recall={recall:.3f} | Precision={precision:.3f}")
```

Most important category variables diagram

```python
from sklearn.inspection import permutation_importance
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Fetch X_test that are transformed
X_test_processed = model.named_steps['preprocess'].transform(X_test)

# Calculate the importance directtly on the classifier model
result = permutation_importance(
    model.named_steps['classifier'],
    X_test_processed,
    y_test,
    n_repeats=10,
    random_state=42
)

# Get the feature name list
feature_names = model.named_steps['preprocess'].get_feature_names_out()

# Merge into DataFrame
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': result.importances_mean
})

# Arrage in decreasing order
importance_df = importance_df.sort_values('Importance', ascending=False).head(15)

# Create a  diagram
plt.figure(figsize=(10,6))
plt.barh(importance_df['Feature'], importance_df['Importance'])
plt.title('Feature Importance (Permutation-based for MLP)')
plt.gca().invert_yaxis()
plt.show()
```
